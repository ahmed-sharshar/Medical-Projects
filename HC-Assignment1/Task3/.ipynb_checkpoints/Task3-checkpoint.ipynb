```python
import cv2
import matplotlib.pyplot as plt
from math import log10, sqrt
import numpy as np
from PIL import Image
from PIL import ImageFilter
```

```python
import pandas as pd
df = pd.read_csv('Frames.txt')
df = df.set_index('FN')
```

```python
for i in filelist:
    for name, row in df.iterrows():
        if i == name:
            print (name)
            cap = cv2.VideoCapture(name)
            first_frame = (df.loc[name][' Frame1'])
            second_frame = (df.loc[name][' Frame2'])
            shot_1 = cap.set(cv2.CAP_PROP_POS_FRAMES, first_frame)
            res1, frame1 = cap.read()
            shot_2 = cap.set(cv2.CAP_PROP_POS_FRAMES, second_frame)
            res2, frame2 = cap.read()
            psnr = cv2.PSNR(frame1, frame2)
            print("Frame1 & Frame2 PSNR  = " , psnr)
```

```python
for i in filelist:
    for name, row in df.iterrows():
        if i == name:
            print (name)
            cap = cv2.VideoCapture(name)
            second_frame = (df.loc[name][' Frame2'])
            shot_2 = cap.set(cv2.CAP_PROP_POS_FRAMES, second_frame)
            res2, frame2 = cap.read()
            image = Image.fromarray(frame2)
            shot_1 = cap.set(cv2.CAP_PROP_POS_FRAMES, first_frame)
            res1, frame1 = cap.read()
            
            gauss = image.filter(ImageFilter.GaussianBlur(radius=3)) #3*3 gauss
            gauss = np.array(gauss)
            psnr = cv2.PSNR(frame1, gauss)
            print("Gaussian Filter PSNR  = " , psnr)
            
            median = image.filter(ImageFilter.MedianFilter(size = 3)) #3*3 median
            median = np.array(median)
            psnr = cv2.PSNR(frame1, median)
            print("Median Filter PSNR  = " , psnr)
            
            box = image.filter(ImageFilter.BoxBlur(1)) #3*3 box
            box = np.array(box)
            psnr = cv2.PSNR(frame1, box)
            print("Box Filter PSNR  = " , psnr)
            
            mode = image.filter(ImageFilter.ModeFilter(size=3)) #mode
            mode = np.array(mode)
            psnr = cv2.PSNR(frame1, mode)
            print("mode Filter PSNR  = " , psnr)
```
