```python
import matplotlib.pyplot as plt
%matplotlib inline
import imageio
import numpy as np
from tkinter import *
from PIL import *
import time
from PIL import ImageTk
import pydicom
```

```python
#Load one image in axial mode
xray = pydicom.read_file("data/IMG0108.dcm")
dicom_img = xray.pixel_array
fig, ax = plt.subplots(1, 1, figsize=(8, 8))
ax.imshow(dicom_img, cmap='gray')
ax.axis('off')
```

```python
def process(x):
    
    #read files as a stack
    vol = imageio.volread("C:/Users/ahmed/Downloads/HC-Asiignment1/Task2/data/", "DICOM")  #please use your path
    n0, n1, n2 = vol.shape
    
    #sperate every view
    axial = []
    for i in range(0,n0-1):
        ax = vol[i,:,:]
        axial.append(ax)
    print(len(axial))

    coronal = []
    for j in range(0,n1-1):
        cor = vol[:,j,:]
        coronal.append(cor)
    print(len(coronal))

    sagittal = []
    for y in range(0,n2-1):
        sig = vol[:,:,y]
        sagittal.append(sig)
    print(len(sagittal))
    
    #make gif and store it based on the button that is clicked on
    if x ==1:
        imgs = axial
        imgs = [Image.fromarray(i) for i in imgs]
        imgs[0].save("axial.gif", save_all=True, append_images=imgs[1:], duration=100, loop=0)
        
    elif x ==2:
        imgs = coronal
        imgs = [Image.fromarray(i) for i in imgs]
        imgs[0].save("coronal.gif", save_all=True, append_images=imgs[1:], duration=100, loop=0)
        
    else:
        imgs = sagittal
        imgs = [Image.fromarray(i) for i in imgs]
        imgs[0].save("sagittal.gif", save_all=True, append_images=imgs[1:], duration=100, loop=0)
        ##########################################################################################

root = Tk()
root.geometry = ("50x50")

#make a function for every view to load the gif and view it

def play_axial():
    process(1)
    global img
    img = Image.open("axial.gif")
    lbl = Label(root)
    lbl.place(x=0 , y=0)
    for img in ImageSequence.Iterator(img):
        img = ImageTk.PhotoImage(img)
        lbl.config(image=img)
        root.update()
        time.sleep(0.1)
    root.after(0,play_axial)

def play_coronal():
    process(2)
    global img
    img = Image.open("coronal.gif")
    lbl = Label(root)
    lbl.place(x=0 , y=0)
    for img in ImageSequence.Iterator(img):
        img = ImageTk.PhotoImage(img)
        lbl.config(image=img)
        root.update()
        time.sleep(0.1)
    root.after(0,play_coronal)

def play_sagittal():
    process(3)
    global img
    img = Image.open("sagittal.gif")
    lbl = Label(root)
    lbl.place(x=0 , y=0)
    for img in ImageSequence.Iterator(img):
        img = ImageTk.PhotoImage(img)
        lbl.config(image=img)
        root.update()
        time.sleep(0.1)
    root.after(0,play_sagittal)
    
def exit():
    root.destroy()
        
        
#button with some boring decorations

Button(root , text = 'Axial' , command =play_axial, height = 1 , width = 7, bg = 'black', fg = 'white'
       , font = "Times 14 bold " ).place(x = 400 , y = 600)

Button(root , text = 'Coronal' , command =play_coronal , height = 1 , width = 7, bg = 'black', fg = 'white'
       , font = "Times 14 bold ").place(x = 300 , y = 600)

Button(root , text = 'Sagittal' , command =play_sagittal, height = 1 , width = 7, bg = 'black', fg = 'white'
       , font = "Times 14 bold " ).place(x = 200 , y = 600)

Button(root , text = 'Stop',  command =exit , height = 1 , width = 7, bg = 'red', fg = 'white'
       , font = "Times 14 bold " ).place(x = 100 , y = 600)

# run it
mainloop()
```
